# Pruning in Transformers



## What is pruning?

Pruning is a model compression technique where we remove weights, neurons, heads, or channels that contribute little to the final prediction.

In transformers, pruning helps make the model smaller and faster while trying to keep accuracy close to the original model.

The main idea is:
- keep the important parts
- remove the less useful parts



## Why pruning is important?

Large transformer models are expensive to run.

Pruning is important because it can:
- reduce model size
- speed up inference
- lower memory usage
- make deployment easier
- cut compute cost

It is useful for mobile apps, edge devices, and production systems.



## Pruning intuition

Imagine a very large model where some attention heads or neurons are rarely useful.

If those parts do not contribute much, removing them may not hurt performance badly.

The model becomes lighter, but still keeps most of the learned knowledge.



## Types of pruning

### 1. Unstructured pruning
Removes individual weights.

### 2. Structured pruning
Removes entire neurons, attention heads, layers, or channels.

### 3. Layer pruning
Removes one or more complete transformer layers.

### 4. Head pruning
Removes attention heads that contribute less.

### 5. Magnitude-based pruning
Removes weights with very small absolute values.

### 6. Dynamic pruning
Prunes during training or inference based on importance.



## How pruning works

A common workflow is:

1. Train a full model.
2. Measure importance of weights, heads, or layers.
3. Remove the least important ones.
4. Fine-tune the pruned model.
5. Evaluate accuracy and speed.

Often, pruning alone can hurt performance a bit, so fine-tuning is important.



## Example

Suppose a transformer has 12 attention heads in a layer.

If analysis shows that 3 heads are rarely useful, those heads can be removed.

This reduces computation while keeping the main behavior of the layer.



## Advantages

- smaller model
- faster inference
- lower memory usage
- sometimes easier deployment
- can be combined with distillation or quantization



## Limitations

- too much pruning can reduce accuracy
- retraining is often needed
- unstructured pruning may not always speed up real hardware execution
- importance estimation can be difficult



## Simple summary

Pruning removes unnecessary parts of a model. In transformers, it is used to reduce size and speed up inference while keeping most of the performance.


```python
import math

# Simple illustration of magnitude-based pruning
weights = [0.82, -0.03, 0.41, 0.01, -0.67, 0.08, 0.002, -0.9]
threshold = 0.05

pruned_weights = [w if abs(w) >= threshold else 0.0 for w in weights]

print("Original weights:", weights)
print("Threshold:", threshold)
print("Pruned weights:  ", pruned_weights)

# Count how many were removed
removed = sum(1 for w in pruned_weights if w == 0.0)
print("Removed weights:", removed)
```
