# Major Encoding Types in NLP
## Comprehensive Guide with Examples, Advantages, Disadvantages & Use Cases

## 1. Integer Encoding
### Purpose: Assign IDs to words

Maps each unique word in vocabulary to a unique integer.

```python
# Integer Encoding Example
from collections import defaultdict

# Create word to ID mapping
text = "the cat sat on the mat and the dog sat on the mat"
words = text.split()

# Build vocabulary
vocab = {}
vocab_idx = 1
for word in set(words):
    if word not in vocab:
        vocab[word] = vocab_idx
        vocab_idx += 1

print("Vocabulary:", vocab)

# Encode text
encoded = [vocab[word] for word in words]
print(f"\nOriginal: {words}")
print(f"Encoded: {encoded}")

# Reverse mapping
idx_to_word = {v: k for k, v in vocab.items()}
decoded = [idx_to_word[idx] for idx in encoded]
print(f"Decoded: {decoded}")
```

```python
# Using sklearn
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
encoded_sklearn = le.fit_transform(words)
print(f"Sklearn Encoded: {encoded_sklearn}")
print(f"Classes: {le.classes_}")
```

### Advantages:
- ✓ Simple and fast to implement
- ✓ Reduces memory footprint
- ✓ Suitable for RNN/LSTM models

### Disadvantages:
- ✗ No semantic meaning preserved
- ✗ Assumes equal distance between integers
- ✗ Cannot capture word relationships
- ✗ Sequence-dependent (order matters)

### Use Cases:
- Input to embedding layers in neural networks
- Machine translation systems
- Text generation models
- RNN/LSTM sequence models

## 2. One-Hot Encoding
### Purpose: Sparse token representation

Creates a binary vector where only one position is 1 (representing the word), rest are 0s.

```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder

# Manual One-Hot Encoding
text = "the cat sat"
words = text.split()
vocab = list(set(words))
vocab_size = len(vocab)

print(f"Vocabulary: {vocab}")
print(f"Vocabulary Size: {vocab_size}\n")

# Create one-hot vectors
one_hot_vectors = {}
for idx, word in enumerate(vocab):
    vector = [0] * vocab_size
    vector[idx] = 1
    one_hot_vectors[word] = vector

for word, vector in one_hot_vectors.items():
    print(f"{word}: {vector}")
```

```python
# Using keras/tensorflow
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical

tokenizer = Tokenizer(num_words=10)
tokenizer.fit_on_texts([text])
sequences = tokenizer.texts_to_sequences([text])[0]

print(f"Sequences: {sequences}")

# One-hot encode
one_hot = to_categorical(sequences, num_classes=10)
print(f"\nOne-Hot Shape: {one_hot.shape}")
print(f"First word one-hot: {one_hot[0]}")
```

### Advantages:
- ✓ Interpretable and simple
- ✓ No assumed ordering between words
- ✓ Works well with small vocabularies
- ✓ Compatible with most ML algorithms

### Disadvantages:
- ✗ Curse of dimensionality (high-dimensional sparse vectors)
- ✗ Memory inefficient for large vocabularies
- ✗ Ignores word semantics and relationships
- ✗ Computationally expensive for large datasets

### Use Cases:
- Categorical variable encoding in traditional ML
- Small vocabulary classification tasks
- Feature engineering for linear models
- Language models with limited vocabulary

## 3. Bag of Words (BoW)
### Purpose: Frequency representation

Represents text as a collection of word counts, ignoring word order and grammar.

```python
from collections import Counter
import pandas as pd

# Manual Bag of Words
documents = [
    "the cat sat on the mat",
    "the dog sat on the log",
    "cats and dogs are pets"
]

# Build vocabulary
all_words = set()
for doc in documents:
    all_words.update(doc.split())

vocab = sorted(list(all_words))
print(f"Vocabulary: {vocab}\n")

# Create BoW representations
bow_vectors = []
for doc in documents:
    word_counts = Counter(doc.split())
    bow_vector = [word_counts.get(word, 0) for word in vocab]
    bow_vectors.append(bow_vector)

# Display as DataFrame
bow_df = pd.DataFrame(bow_vectors, columns=vocab, index=[f"Doc{i+1}" for i in range(len(documents))])
print(bow_df)
```

```python
# Using sklearn CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
bow_sklearn = cv.fit_transform(documents)

print("Feature names:", cv.get_feature_names_out())
print("\nBoW Matrix (dense):")
print(bow_sklearn.toarray())
print(f"\nBoW Matrix Shape: {bow_sklearn.shape}")
```

### Advantages:
- ✓ Simple to understand and implement
- ✓ Fast to compute
- ✓ Works well with text classification
- ✓ Captures word importance through frequency
- ✓ Sparse representation (memory efficient)

### Disadvantages:
- ✗ Loses word order (sequence information)
- ✗ Treats all words equally regardless of importance
- ✗ High-dimensional sparse vectors for large vocabularies
- ✗ Cannot capture semantic meaning
- ✗ Doesn't handle synonyms

### Use Cases:
- Email spam detection
- Text classification and sentiment analysis
- Document similarity
- Information retrieval
- Naive Bayes classification

## 4. TF-IDF (Term Frequency-Inverse Document Frequency)
### Purpose: Importance-weighted frequencies

Weighs word frequencies by their importance across documents.

```python
import math

# Manual TF-IDF calculation
documents = [
    "machine learning is great",
    "deep learning is powerful",
    "machine learning and deep learning"
]

# Build vocabulary
all_words = set()
for doc in documents:
    all_words.update(doc.split())

vocab = sorted(list(all_words))

# Calculate IDF
def calculate_idf(documents, word, vocab):
    doc_count = sum(1 for doc in documents if word in doc.split())
    if doc_count == 0:
        return 0
    return math.log(len(documents) / doc_count)

# Calculate TF-IDF for each document
tfidf_vectors = []
for doc in documents:
    words = doc.split()
    tfidf_vector = []
    for word in vocab:
        tf = words.count(word) / len(words)
        idf = calculate_idf(documents, word, vocab)
        tfidf = tf * idf
        tfidf_vector.append(tfidf)
    tfidf_vectors.append(tfidf_vector)

tfidf_df = pd.DataFrame(tfidf_vectors, columns=vocab, index=[f"Doc{i+1}" for i in range(len(documents))])
print(tfidf_df.round(4))
```

```python
# Using sklearn TfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_sklearn = tfidf.fit_transform(documents)

print("Feature names:", tfidf.get_feature_names_out())
print("\nTF-IDF Matrix (dense):")
print(tfidf_sklearn.toarray().round(4))
print(f"\nTF-IDF Matrix Shape: {tfidf_sklearn.shape}")
```

### Advantages:
- ✓ Reduces importance of common words (stop words)
- ✓ Highlights important and rare words
- ✓ Better than BoW for most tasks
- ✓ Computationally efficient
- ✓ Works well with sparse representations

### Disadvantages:
- ✗ Still loses word order and context
- ✗ Cannot capture word relationships
- ✗ Doesn't work well for short documents
- ✗ May penalize domain-specific terms
- ✗ Ignores semantic meaning

### Use Cases:
- Information retrieval and search engines
- Document similarity and clustering
- Text classification with feature weighting
- Keyword extraction
- Recommender systems for content

## 5. Word Embeddings
### Purpose: Semantic vectors

Learns dense vector representations where semantically similar words have similar vectors.

```python
# Using Gensim Word2Vec
from gensim.models import Word2Vec
import numpy as np

# Prepare sentences
sentences = [
    "the king and the queen",
    "the man and the woman",
    "a man is driving a car",
    "a woman is driving a car",
    "the king is wise and strong"
]

# Tokenize
tokenized = [sent.split() for sent in sentences]

# Train Word2Vec
w2v = Word2Vec(sentences=tokenized, vector_size=100, window=3, min_count=1, sg=1)

print(f"Word 'king' embedding shape: {w2v.wv['king'].shape}")
print(f"Word 'king' embedding (first 10 dims): {w2v.wv['king'][:10]}")

# Find similar words
print(f"\nWords similar to 'king': {w2v.wv.most_similar('king', topn=3)}")
print(f"Words similar to 'woman': {w2v.wv.most_similar('woman', topn=3)}")

# Word analogy: king - man + woman = queen
try:
    result = w2v.wv.most_similar(positive=['king', 'woman'], negative=['man'], topn=3)
    print(f"\nAnalogy (king - man + woman): {result}")
except:
    print("\nInsufficient data for analogy")
```

```python
# Using pre-trained GloVe embeddings (fastText as alternative)
# Download pre-trained word2vec model
from gensim.models import KeyedVectors

# Example: compute similarity between words
similarity = w2v.wv.similarity('king', 'queen')
print(f"Similarity between 'king' and 'queen': {similarity:.4f}")

similarity = w2v.wv.similarity('king', 'car')
print(f"Similarity between 'king' and 'car': {similarity:.4f}")

# Get vector for a word
vector = w2v.wv['king']
print(f"\nVector representation of 'king': {vector}")
```

### Advantages:
- ✓ Captures semantic meaning and relationships
- ✓ Dense, low-dimensional vectors
- ✓ Enables analogy solving (king - man + woman = queen)
- ✓ Memory efficient compared to sparse methods
- ✓ Pre-trained models available (Word2Vec, GloVe, fastText)
- ✓ Excellent for downstream tasks

### Disadvantages:
- ✗ Context-independent (same representation for all uses)
- ✗ Requires large corpus for training
- ✗ May inherit biases from training data
- ✗ Sensitive to hyperparameters
- ✗ Black-box nature (hard to interpret)

### Use Cases:
- Word similarity and analogy tasks
- Feature input for NLP models
- Document clustering and classification
- Semantic search
- Recommendation systems
- Named entity recognition
- Machine translation

## 6. Positional Encoding
### Purpose: Sequence order representation

Encodes position information in sequences using sinusoidal functions (used in Transformers).

```python
import numpy as np
import matplotlib.pyplot as plt

def positional_encoding(seq_length, d_model):
    """
    Generate positional encoding matrix
    pos: position in sequence
    i: dimension index
    PE(pos, 2i) = sin(pos / 10000^(2i/d_model))
    PE(pos, 2i+1) = cos(pos / 10000^(2i/d_model))
    """
    pe = np.zeros((seq_length, d_model))
    
    for pos in range(seq_length):
        for i in range(0, d_model, 2):
            angle_rates = 1 / np.power(10000, (2 * (i//2)) / np.float32(d_model))
            pe[pos, i] = np.sin(pos * angle_rates)
            if i + 1 < d_model:
                pe[pos, i+1] = np.cos(pos * angle_rates)
    
    return pe

# Generate positional encoding
seq_length = 100
d_model = 512
pe = positional_encoding(seq_length, d_model)

print(f"Positional Encoding Shape: {pe.shape}")
print(f"\nFirst 3 positions, first 8 dimensions:")
print(pe[:3, :8])

# Visualize
plt.figure(figsize=(12, 4))
plt.imshow(pe[:, :50], cmap='viridis', aspect='auto')
plt.colorbar(label='Encoding Value')
plt.xlabel('Dimension')
plt.ylabel('Position')
plt.title('Positional Encoding Heatmap (first 50 dimensions)')
plt.tight_layout()
plt.show()
```

```python
# Using PyTorch implementation
import torch

class PositionalEncoding(torch.nn.Module):
    def __init__(self, d_model, max_len=5000):
        super(PositionalEncoding, self).__init__()
        
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * 
                              -(math.log(10000.0) / d_model))
        
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        
        self.register_buffer('pe', pe.unsqueeze(0))
    
    def forward(self, x):
        return x + self.pe[:, :x.size(1)]

import math
pe_layer = PositionalEncoding(d_model=512, max_len=100)
print(f"PyTorch PE shape: {pe_layer.pe.shape}")
print(f"First position encoding (first 8 dims): {pe_layer.pe[0, 0, :8]}")
```

### Advantages:
- ✓ Preserves sequence order information
- ✓ Differentiable and learnable (in some variants)
- ✓ Works well with parallel processing (Transformers)
- ✓ Can generalize to longer sequences than training
- ✓ Lightweight (mathematical function, not learned)

### Disadvantages:
- ✗ Fixed encoding pattern (limited adaptability)
- ✗ May struggle with very long sequences
- ✗ Arbitrary choice of sinusoidal functions
- ✗ Added computational overhead

### Use Cases:
- Transformer models (BERT, GPT, T5)
- Sequence-to-sequence models
- Attention mechanisms
- Machine translation
- Attention-based time series models

## 7. Contextual Encoding
### Purpose: Dynamic meaning based on context

Generates different representations for the same word depending on context (e.g., ELMo, BERT).

```python
# Using Hugging Face Transformers - BERT
from transformers import AutoTokenizer, AutoModel
import torch

# Load pre-trained model and tokenizer
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name, output_hidden_states=True)

# Example sentences
sentences = [
    "The bank is by the river.",
    "I deposited money in the bank."
]

# Encode and get contextual embeddings
for sentence in sentences:
    inputs = tokenizer.encode(sentence, return_tensors='pt')
    outputs = model(inputs)
    hidden_states = outputs[2]
    
    # Get embedding for 'bank' token
    tokens = tokenizer.convert_ids_to_tokens(inputs[0])
    bank_idx = tokens.index('bank')
    bank_embedding = hidden_states[-1][0, bank_idx, :10]
    
    print(f"Sentence: {sentence}")
    print(f"'bank' embedding (first 10 dims): {bank_embedding}")
    print()
```

```python
# Compare similarity of 'bank' in different contexts
embeddings = []
for sentence in sentences:
    inputs = tokenizer.encode(sentence, return_tensors='pt')
    with torch.no_grad():
        outputs = model(inputs)
    hidden_states = outputs[2][-1]  # Last hidden state
    tokens = tokenizer.convert_ids_to_tokens(inputs[0])
    bank_idx = tokens.index('bank')
    embeddings.append(hidden_states[0, bank_idx, :].numpy())

# Calculate cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity([embeddings[0]], [embeddings[1]])[0][0]
print(f"Cosine similarity of 'bank' embeddings: {similarity:.4f}")
print("\nNote: While same word, contexts are different (financial vs geographical)")
```

### Advantages:
- ✓ Captures context and polysemy (multiple meanings)
- ✓ Superior performance on downstream tasks
- ✓ Bidirectional context (in BERT)
- ✓ Transfer learning capabilities
- ✓ Pre-trained models available
- ✓ Handles out-of-vocabulary words well

### Disadvantages:
- ✗ Computationally expensive
- ✗ Requires fine-tuning for specific tasks
- ✗ May introduce bias from training data
- ✗ Less interpretable than simpler methods
- ✗ Requires GPU for reasonable speed

### Use Cases:
- Sentiment analysis
- Named entity recognition
- Text classification
- Question answering
- Machine translation
- Semantic similarity
- Intent detection in chatbots

## 8. Sentence Embeddings
### Purpose: Whole sentence vectors

Creates embeddings for entire sentences or documents.

```python
# Using Sentence-BERT
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Load model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Example sentences
sentences = [
    "The cat is sitting on the mat.",
    "A feline is resting on a rug.",
    "The weather is sunny today.",
    "Dogs are loyal animals."
]

# Encode sentences
embeddings = model.encode(sentences)

print(f"Sentence Embedding Shape: {embeddings.shape}")
print(f"First sentence embedding (first 10 dims): {embeddings[0, :10]}")

# Calculate similarities
print("\nSentence Similarities:")
for i in range(len(sentences)):
    for j in range(i+1, len(sentences)):
        sim = cosine_similarity([embeddings[i]], [embeddings[j]])[0][0]
        print(f"Similarity({i+1}, {j+1}): {sim:.4f}")
```

```python
# Document similarity / clustering
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Cluster sentences
kmeans = KMeans(n_clusters=2, random_state=42)
labels = kmeans.fit_predict(embeddings)

print("Cluster assignments:")
for i, (sent, label) in enumerate(zip(sentences, labels)):
    print(f"Cluster {label}: {sent}")

# Visualize with PCA
pca = PCA(n_components=2)
embeddings_2d = pca.fit_transform(embeddings)

plt.figure(figsize=(8, 6))
plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1], c=labels, cmap='viridis', s=100)
for i, sent in enumerate(sentences):
    plt.annotate(f"S{i+1}", (embeddings_2d[i, 0], embeddings_2d[i, 1]))
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Sentence Embeddings Visualization')
plt.tight_layout()
plt.show()
```

### Advantages:
- ✓ Captures semantic meaning of entire documents
- ✓ Fixed-size vectors (easy to compare)
- ✓ Excellent for document similarity and clustering
- ✓ Pre-trained models available
- ✓ Enables semantic search
- ✓ Better than averaging word embeddings

### Disadvantages:
- ✗ Computationally expensive for large documents
- ✗ May lose fine-grained details
- ✗ Sensitive to model choice and training data
- ✗ Requires significant computation for fine-tuning

### Use Cases:
- Document clustering and classification
- Semantic search engines
- Duplicate detection
- Document similarity assessment
- Recommendation systems
- Paraphrase detection
- Information retrieval

## 9. Character Encoding
### Purpose: Character-level NLP

Represents text at the character level instead of word level.

```python
# Manual Character Encoding
text = "hello world"

# Build character vocabulary
char_set = sorted(list(set(text)))
char_to_idx = {c: i for i, c in enumerate(char_set)}
idx_to_char = {i: c for c, i in char_to_idx.items()}

print(f"Character Set: {char_set}")
print(f"Char to Index: {char_to_idx}\n")

# Encode text
encoded = [char_to_idx[c] for c in text]
print(f"Original: {text}")
print(f"Encoded: {encoded}")

# Decode
decoded = ''.join([idx_to_char[i] for i in encoded])
print(f"Decoded: {decoded}")
```

```python
# Character one-hot encoding
import numpy as np

vocab_size = len(char_set)
char_onehot = {}

for i, char in enumerate(char_set):
    vector = np.zeros(vocab_size)
    vector[i] = 1
    char_onehot[char] = vector

print(f"One-hot encoded 'h': {char_onehot['h']}")
print(f"One-hot encoded 'e': {char_onehot['e']}")
print(f"One-hot encoded ' ': {char_onehot[' ']}")

# Create sequence of one-hot vectors
sequence_onehot = np.array([char_onehot[c] for c in "hello"])
print(f"\nSequence shape for 'hello': {sequence_onehot.shape}")
```

```python
# Character embeddings with neural network
import torch
import torch.nn as nn

class CharEmbedding(nn.Module):
    def __init__(self, vocab_size, embedding_dim):
        super(CharEmbedding, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
    
    def forward(self, text_indices):
        return self.embedding(text_indices)

# Example
vocab_size = len(char_set)
embedding_dim = 50
char_emb_model = CharEmbedding(vocab_size, embedding_dim)

# Convert text to indices
text_indices = torch.tensor([char_to_idx[c] for c in "hello"])
embeddings = char_emb_model(text_indices)

print(f"Character embeddings shape: {embeddings.shape}")
print(f"'h' embedding (first 5 dims): {embeddings[0, :5]}")
```

### Advantages:
- ✓ Handles out-of-vocabulary words (OOV)
- ✓ Smaller vocabulary size
- ✓ Can capture morphological patterns (prefixes, suffixes)
- ✓ Works for languages with large alphabets
- ✓ Handles misspellings better than word-level

### Disadvantages:
- ✗ Loses high-level semantic information
- ✗ Sequences become very long
- ✗ More computationally expensive
- ✗ Harder to capture word-level patterns
- ✗ Slower to train due to longer sequences

### Use Cases:
- Language modeling (especially character-level)
- Handling morphologically rich languages
- Spell checking and correction
- Name recognition in low-resource languages
- Code analysis and programming language tasks
- Text generation at character level

## 10. Byte Encoding
### Purpose: Raw byte representation

Represents text as raw byte values (0-255).

```python
# Manual Byte Encoding
text = "hello"

# Convert to bytes
byte_encoded = text.encode('utf-8')
print(f"Text: {text}")
print(f"Byte encoded: {byte_encoded}")
print(f"Byte values: {list(byte_encoded)}")

# Decode back
decoded = byte_encoded.decode('utf-8')
print(f"Decoded: {decoded}")

# Character to byte mapping
print("\nCharacter to Byte mapping:")
for char in text:
    byte_val = char.encode('utf-8')[0]
    print(f"'{char}': {byte_val}")
```

```python
# Byte-Pair Encoding (BPE) - commonly used in modern NLP
# This is a simplification of BPE concept

class SimpleBPE:
    def __init__(self, vocab_size=100):
        self.vocab_size = vocab_size
        self.vocab = {}
    
    def train(self, texts):
        # Build initial byte vocabulary
        for i in range(256):
            self.vocab[bytes([i])] = i
        
    def encode(self, text):
        return list(text.encode('utf-8'))
    
    def decode(self, byte_seq):
        return bytes(byte_seq).decode('utf-8', errors='ignore')

bpe = SimpleBPE()
bpe.train([])

texts = ["hello world", "python programming"]
for text in texts:
    encoded = bpe.encode(text)
    decoded = bpe.decode(encoded)
    print(f"Text: {text}")
    print(f"Byte encoded: {encoded}")
    print(f"Decoded: {decoded}\n")
```

```python
# Byte embeddings for neural networks
import torch
import torch.nn as nn

class ByteEmbedding(nn.Module):
    def __init__(self, embedding_dim=32):
        super(ByteEmbedding, self).__init__()
        # 256 possible byte values
        self.embedding = nn.Embedding(256, embedding_dim)
    
    def forward(self, byte_sequence):
        return self.embedding(byte_sequence)

# Example
byte_emb_model = ByteEmbedding(embedding_dim=32)

# Encode text to bytes
text = "hello"
byte_indices = torch.tensor(list(text.encode('utf-8')))

# Get embeddings
embeddings = byte_emb_model(byte_indices)

print(f"Input text: {text}")
print(f"Byte indices: {byte_indices.tolist()}")
print(f"Embeddings shape: {embeddings.shape}")
print(f"'h' embedding (first 5 dims): {embeddings[0, :5]}")
```

### Advantages:
- ✓ Language agnostic (works for any encoding)
- ✓ Handles multilingual text seamlessly
- ✓ Very small fixed vocabulary (256 bytes)
- ✓ No OOV problems
- ✓ Robust to noise and corrupted text

### Disadvantages:
- ✗ Very long sequences (especially for multibyte characters)
- ✗ Computationally expensive
- ✗ Difficult to capture semantic meaning
- ✗ Inefficient for most downstream tasks
- ✗ Hard to interpret

### Use Cases:
- Multilingual NLP systems
- Byte-Pair Encoding (BPE) for transformer models
- Low-level text processing
- Handling arbitrary binary data
- Robustness to Unicode and encoding issues
- Modern tokenization (used in GPT models)

## When to Use Each Encoding Type

| Scenario | Recommended Encoding |
|----------|---------------------|
| Baseline classification | Bag of Words / TF-IDF |
| Modern deep learning | Contextual Embeddings (BERT, etc.) |
| Quick similarity search | Sentence Embeddings |
| Word relationships | Word Embeddings (Word2Vec, GloVe) |
| Transformer models | Positional + Contextual |
| Multilingual tasks | Byte Encoding |
| Morphologically rich languages | Character Encoding |
| Sequence models (RNN/LSTM) | Integer Encoding |
| Feature engineering (traditional ML) | One-Hot / BoW / TF-IDF |
| Production systems | Word/Contextual Embeddings or Sentence Embeddings |

## Key Takeaways

1. **Simple encodings** (Integer, One-Hot, BoW) are fast but lose semantic information
2. **TF-IDF** improves on BoW by weighting word importance
3. **Word Embeddings** capture semantic meaning but are context-independent
4. **Contextual Encodings** (BERT, ELMo) are state-of-the-art but computationally expensive
5. **Sentence Embeddings** work well for document-level tasks
6. **Character/Byte Encoding** handles edge cases (OOV, multilingual, misspellings)
7. **Positional Encoding** is essential for attention-based models
8. Choose encoding based on task requirements, computational budget, and data availability
