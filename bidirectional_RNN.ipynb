# Bidirectional RNN — How It Works

This notebook explains how a Bidirectional RNN works, step by step, in one complete flow.

## 1. What is a Bidirectional RNN?

A Bidirectional Recurrent Neural Network (BiRNN) is an improved version of a normal RNN.
It processes the same sequence in **two directions**:

- **Forward**: left to right
- **Backward**: right to left

This lets the model use both past and future context for each token.

## 2. Why Normal RNNs Are Limited

A normal RNN reads a sequence one token at a time from left to right.

Example sentence:

`The movie was not good`

When the model reaches the word `not`, it has only seen the words before it, not the words after it.
So it may miss important future context.

## 3. The Core Idea of BiRNN

A BiRNN uses **two RNNs**:

- one forward RNN that reads from the beginning to the end
- one backward RNN that reads from the end to the beginning

Then it combines the hidden states from both directions.

## 4. Forward Pass

The forward RNN processes the sequence like this:

```text
The → cat → sat
```

At each time step, it uses:

- the current input token
- the previous hidden state

Normal RNN hidden state formula:

$$
h_t = f(W_x x_t + W_h h_{t-1} + b)
$$

## 5. Backward Pass

The backward RNN processes the same sequence in reverse:

```text
sat → cat → The
```

This gives the model future context for each token.

## 6. Combining the Two Directions

For each token, the forward hidden state and backward hidden state are combined.

A common way is concatenation:

$$
h_t = [\overrightarrow{h_t}; \overleftarrow{h_t}]
$$

This final representation contains both past and future information. 

## 7. Full Example

Sentence:

`The movie was not good`

### Forward RNN
Reads left to right:

```text
The → movie → was → not → good
```

At the word `not`, it knows the left context:

`The movie was`

### Backward RNN
Reads right to left:

```text
good → not → was → movie → The
```

At the word `not`, it already knows the right context:

`good`

### Combined Representation
The token `not` now has access to both sides, which helps the model understand the sentence better.

## 8. Why BiRNN Helps

Language often depends on both sides of a word.

Example:

`He went to the bank`

The meaning of `bank` may depend on the words before and after it.
A BiRNN captures this better than a one-way RNN.

## 9. Training Flow

A BiRNN is trained like other neural networks:

```text
Input
↓
Forward RNN
↓
Backward RNN
↓
Combine outputs
↓
Loss calculation
↓
Backpropagation
↓
Weight update
```

## 10. BiRNN Variants

The most common bidirectional versions are:

- **BiLSTM**: Bidirectional LSTM
- **BiGRU**: Bidirectional GRU

These are usually more powerful than a plain BiRNN because they handle long dependencies better.

## 11. Where BiRNNs Are Used

BiRNNs were widely used in NLP tasks such as:

- sentiment analysis
- named entity recognition
- speech recognition
- sequence labeling
- machine translation

## 12. Limitations

Even though BiRNNs are better than normal RNNs, they still have some issues:

- they are sequential
- they are slower to train
- they are hard to parallelize
- they can struggle with very long dependencies

That is one reason why transformers became the dominant architecture in modern NLP.

## 13. BiRNN vs Transformer

| Feature | BiRNN | Transformer |
|---|---|---|
| Direction | Forward + backward | Full attention |
| Future context | Yes | Yes |
| Parallel training | No | Yes |
| Speed | Slower | Faster |
| Long-range dependencies | Better than RNN | Excellent |

## 14. Simple Intuition

A normal RNN reads a sentence in one direction.

A Bidirectional RNN reads both directions and merges the information, so each word gets a richer understanding of context.

## 15. One-Line Summary

A Bidirectional RNN works by running one RNN forward and another backward over the same sequence, then combining both hidden states so each token can use past and future context.

````python
# Conceptual example: forward and backward context for a token
sentence = ['The', 'movie', 'was', 'not', 'good']
target_index = 3  # 'not'

forward_context = sentence[:target_index]
backward_context = sentence[target_index + 1:]

print('Sentence:', ' '.join(sentence))
print('Target token:', sentence[target_index])
print('Forward context:', forward_context)
print('Backward context:', backward_context)
````
