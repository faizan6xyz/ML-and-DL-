# Quantization in Transformers



## What is quantization?

Quantization is a model compression technique that reduces the numerical precision of weights and activations.

For example, a model that uses 32-bit floating point values (FP32) may be converted to 16-bit, 8-bit, or even lower precision.

The goal is to make the model smaller, faster, and cheaper to run.



## Why quantization is important?

Transformers often have millions or billions of parameters.

Quantization helps because it can:
- reduce memory usage
- speed up inference
- lower storage size
- improve deployment on limited hardware
- reduce bandwidth cost



## Quantization intuition

A floating-point number stores more precision than is often needed at inference time.

Quantization replaces high-precision numbers with lower-precision approximations.

This usually causes a small loss in accuracy, but the savings can be large.



## Types of quantization

### 1. Post-training quantization (PTQ)
Quantize a trained model after training is finished.

### 2. Quantization-aware training (QAT)
Train the model while simulating quantization effects.

### 3. Dynamic quantization
Quantize weights and compute activations dynamically.

### 4. Static quantization
Use calibration data to quantize both weights and activations.

### 5. Weight-only quantization
Only weights are quantized.

### 6. Mixed-precision quantization
Different parts of the model use different precisions.



## How quantization works

A typical process is:

1. Start with a trained model.
2. Choose the target precision, such as INT8 or FP16.
3. Convert weights and possibly activations.
4. Calibrate or fine-tune if needed.
5. Measure the accuracy-speed tradeoff.



## Example

Suppose a weight value is 0.873421 in FP32.

After INT8 quantization, it may be approximated by a smaller integer representation plus a scale factor.

That approximate value uses much less memory and is faster to compute with.



## Advantages

- smaller model size
- faster inference
- lower memory usage
- better hardware efficiency
- useful for deployment on GPUs, CPUs, and edge devices



## Limitations

- may slightly reduce accuracy
- very low precision can hurt quality
- some models are more sensitive than others
- calibration or fine-tuning may be needed



## Simple summary

Quantization reduces numerical precision to make transformer models smaller and faster. It is one of the most common ways to optimize deployment.


```python
import numpy as np

# Simple illustration of symmetric INT8-style quantization
values = np.array([0.12, -0.55, 0.91, -0.08, 0.33], dtype=np.float32)

max_abs = np.max(np.abs(values))
scale = max_abs / 127.0 if max_abs != 0 else 1.0

q_values = np.round(values / scale).astype(np.int8)
dequantized = q_values.astype(np.float32) * scale

print("Original values:   ", values)
print("Scale factor:      ", scale)
print("Quantized int8:    ", q_values)
print("Dequantized values:", np.round(dequantized, 4))
```
