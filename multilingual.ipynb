# What Makes a Transformer Multilingual?

The architecture is mostly the same. The main differences are:

1. Shared multilingual tokenizer
2. Shared embeddings across languages
3. Training on many languages together
4. Cross-lingual representation learning

## 1. Shared Multilingual Tokenizer

This is the first major change.

A normal English Transformer may use a vocabulary like `cat`, `dog`, `playing`, `beautiful`. It only understands English tokens.

A multilingual Transformer uses one tokenizer trained on many languages together: English, Hindi, French, Arabic, Chinese, Spanish, and more.

It often works with subwords or byte-level pieces, which helps it handle rare words, multiple scripts, and mixed-language text.

Example:

- `I am learning` → `['I', 'am', 'learn', '##ing']`
- `मैं सीख रहा हूँ` → `['मैं', 'सीख', 'रहा', 'हूँ']`

## 2. Shared Embeddings Across Languages

After tokenization, tokens become vectors.

In a normal Transformer, English words live in an English-only embedding space.

In a multilingual Transformer, words from many languages live in the same embedding space. Words with similar meanings, such as `cat`, `chat`, and `बिल्ली`, end up near each other in vector space.

This creates semantic alignment across languages.

## 3. Training on Many Languages Together

A normal Transformer is trained on one language or one main domain.

A multilingual Transformer is trained on text from many languages together. The same neural network weights are shared across all languages, so English, Hindi, French, and other languages update the same model parameters.

That shared training helps the model learn language patterns that are common across languages.

## 4. Cross-Lingual Representation Learning

This is the most important idea.

Cross-lingual means learning across languages. The model learns that different sentences in different languages can express the same meaning.

Example:

- `I love this movie` → positive sentiment
- `मुझे यह फिल्म पसंद है` → also positive sentiment

Even if the model has not seen many labeled Hindi examples, it can transfer what it learned from English because both sentences map to similar internal representations.

This is called cross-lingual transfer.

## Final Big Picture

A normal Transformer:

- one language
- one vocabulary
- one representation space

A multilingual Transformer:

- many languages
- shared tokenizer
- shared embeddings
- shared neural network
- shared semantic space

So the main idea is simple: the model learns that different words in different languages can point to the same concepts.
