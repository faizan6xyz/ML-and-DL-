## working of the Neutral network 

Neutral Network has 3 layers :

    1. Input Layer
        Takes the input form the user 

    2. Hidden Layer 
            Hidden Layer Operations for Each neuron:
                1.Receives inputs
                2.Multiplies by weights
                3.Adds bias
                4.Applies activation function
                5.Sends output forward

    3. output Layer
        gives the final prediction/result (perform the calculation)


A neuron: Takes inputs , Multiplies by weights , Adds bias , Applies activation function
(inputs have weights and the neuron has the bias )

Step 1 — Input Layer

        The network receives input values.

        Example:
            Hours studied = 5
            Attendance = 80



Step 2 — Multiply by Weights

        Each input gets multiplied by weights.

        Weights decide importance.

        Example:
            z = w1x1 +w2x2 +b

        Suppose:
            study hours matter more
            attendance matters less

        Then weight for study hours becomes larger.

Step 3 — Add Bias

    Bias shifts the output slightly.
    It helps the model fit data better.

Step 4 — Activation Function

    An Activation Function is a mathematical function used in a neural network neuron to decide: how much information should pass to the next layer. It adds non-linearity to the neural network.
    Without activation functions: neural networks would behave like simple linear models and could not learn complex patterns.

    The neuron output passes through an activation function.

    Example ReLU:
        f(x)=max(0,x)

    Why activation functions?
    Because real-world data is nonlinear.

    Without activation:
        neural network becomes just linear math.

    Types of activation function :
        ReLU : Outputs the input if positive, otherwise outputs 0.
        Sigmoid : Converts values into probabilities between 0 and 1.
        Tanh : Converts values into range -1 to 1 with zero-centered output. 
        Softmax : Converts outputs into probability distribution for multi-class classification. 
        Leaky ReLU  :  Like ReLU, but allows small negative values instead of zero. 
        ELU : Gives smooth negative outputs to improve learning stability. 
        Swish : Multiplies input with sigmoid for smoother learning. 
        GELU : Smoothly selects important inputs, commonly used in Transformers. 
        Linear : Outputs the same input value without modification. 

Step 5 — Hidden Layers Process Information

        Outputs from one layer become inputs to next layer.

        Example:

        Input Layer → Hidden Layer 1 → Hidden Layer 2 → Output

        Each layer learns different patterns.

        Example in image recognition:

        First layer → edges
        Next layer → shapes
        Final layer → objects

Step 6 — Output Layer Gives Prediction

        Final layer produces result.

        Examples:

        probability of passing
        cat/dog classification
        stock price prediction

        For classification:
        Softmax is often used.

Step 7 — Loss Function Measures Error

        The model checks:
        “How wrong was I?”

        Loss functions calculate prediction error.

        Example:

        predicted = Pass
        actual = Fail

        Large loss = bad prediction.

Step 8 — Backpropagation

        This is the main learning step.

        The network:

        sends error backward
        calculates which weights caused error
        adjusts them

        This process is called:

        Backpropagation

Step 9 — Gradient Descent Updates Weights
        Weights are updated using gradients.

        Update rule:
            w=w−η(∂L/∂w)
            Where:
                w = weight
                L = loss
                η = learning rate

        Goal:
        reduce loss gradually.

Step 10 — Repeat for Many Epochs

        The process repeats again and again.
        Forward → Error → Backward → Update

        Eventually:
            predictions improve
            loss decreases

        This is learning.

# types of Neural Network 
1. Perceptron

        The simplest neural network.

        Structure:
            Input → Output

        Working

        The perceptron:
            takes inputs
            multiplies by weights
            adds bias
            applies activation

        Equation:

        Features
            single-layer network
            binary classification
            linear decision boundary

        Advantages
            simple
            fast
            easy to understand

        Disadvantages
            only solves linearly separable problems
            cannot solve XOR problem

        Applications
            simple classification
            basic ML learning

2. Multilayer Perceptron (MLP)

        Neural network with multiple hidden layers.

        Structure:
            Input → Hidden Layer(s) → Output

        Working

        The MLP:
            performs forward propagation
            calculates loss
            uses backpropagation
            updates weights using gradient descent

        Features
            multiple hidden layers
            nonlinear learning
            fully connected architecture

        Advantages
            learns complex patterns
            powerful and flexible
            handles nonlinear data

        Disadvantages
            computationally expensive
            prone to overfitting
            difficult hyperparameter tuning

        Applications
            regression
            classification
            healthcare prediction
            finance

3. Feedforward Neural Network (FNN)

        Neural network where data moves only forward.

        Structure:
            Input → Hidden → Output

        Working

        The network:
            sends data layer by layer
            performs no feedback loops
            computes outputs directly

        Features
            one-directional data flow
            simple architecture
            foundation of many NNs

        Advantages
            easy implementation
            stable learning
            fast inference

        Disadvantages
            no memory
            poor for sequential data

        Applications
            classification
            regression
            pattern recognition

4. Convolutional Neural Network (CNN)

        Neural network mainly used for image processing.

        Structure:
            Image → Convolution → Pooling → Fully Connected → Output

        Working

        CNN:
            applies filters/kernels
            extracts image features
            reduces dimensions using pooling
            performs classification

        Features
            automatic feature extraction
            spatial learning
            parameter sharing

        Advantages
            excellent for images
            fewer parameters than MLP for images
            high accuracy in vision tasks

        Disadvantages
            high computational cost
            requires large datasets
            slow training

        Applications
            facial recognition
            object detection
            medical imaging
            self-driving cars

5. Recurrent Neural Network (RNN)

        Neural network designed for sequential data.

        Structure:
            Input → Hidden State → Output
                    ↑
                Previous State

        Working

        RNN:
            remembers previous inputs
            passes hidden states forward
            learns sequence relationships

        Features
            memory mechanism
            sequential learning
            temporal dependency handling

        Advantages
            good for sequence data
            handles time-series information

        Disadvantages
            vanishing gradient problem
            weak long-term memory
            difficult training

        Applications
            speech recognition
            language modeling
            text generation
            time-series forecasting

6. Long Short-Term Memory (LSTM)

        Advanced version of RNN with memory gates.

        Structure:
            Input → LSTM Cell → Output

        Working

        LSTM:
            uses memory cells
            controls information using gates
            remembers long-term dependencies

        Gates:
            Forget Gate
            Input Gate
            Output Gate

        Features
            long-term memory
            gate-based architecture
            solves vanishing gradient problem

        Advantages
            strong sequence learning
            excellent long-term memory
            better than vanilla RNN

        Disadvantages
            slow training
            complex architecture
            computationally heavy

        Applications
            chatbots
            translation systems
            NLP
            speech systems

7. Gated Recurrent Unit (GRU)

        Simplified version of LSTM.

        Structure:
            Input → GRU Cell → Output

        Working

        GRU:
            uses update and reset gates
            combines memory and hidden state
            simplifies sequence learning

        Features
            fewer gates
            faster than LSTM
            efficient architecture

        Advantages
            faster training
            fewer parameters
            efficient performance

        Disadvantages
            sometimes less accurate than LSTM
            less expressive memory control

        Applications
            NLP
            speech recognition
            sequence prediction

8. Autoencoder

        Neural network for compression and reconstruction.

        Structure:
            Input → Encoder → Bottleneck → Decoder → Output

        Working

        Autoencoder:
            compresses input data
            learns hidden representation
            reconstructs original input

        Features
            unsupervised learning
            dimensionality reduction
            feature extraction

        Advantages
            good compression
            anomaly detection
            feature learning

        Disadvantages
            reconstruction errors
            difficult tuning

        Applications
            image denoising
            recommendation systems
            anomaly detection

9. Generative Adversarial Network (GAN)

        Neural network for generating realistic data.

        Structure:
            Generator ↔ Discriminator

        Working

        Generator:
            creates fake data

        Discriminator:
            checks whether data is real or fake

        Both networks:
            compete and improve together

        Features
            generative learning
            adversarial training
            realistic output generation

        Advantages
            realistic image generation
            high-quality synthetic data

        Disadvantages
            unstable training
            mode collapse
            difficult optimization

        Applications
            AI art
            deepfakes
            image synthesis

10. Transformer

        Modern neural network architecture used in LLMs.

        Structure:
            Input → Self-Attention → Feedforward Layers → Output

        Working

        Transformer:
            processes all inputs simultaneously
            uses self-attention mechanism
            learns contextual relationships

        Features
            parallel processing
            attention mechanism
            long-context understanding

        Advantages
            state-of-the-art NLP
            very scalable
            excellent context learning

        Disadvantages
            massive computation cost
            expensive training
            requires huge datasets

        Applications
            ChatGPT
            machine translation
            code generation
            generative AI

11. Graph Neural Network (GNN)

        Neural network designed for graph data.

        Structure:
            Nodes ↔ Edges ↔ Message Passing

        Working

        GNN:
            exchanges information between nodes
            learns relationships in graphs
            updates node representations

        Features
            graph-based learning
            relational understanding
            node interaction learning

        Advantages
            excellent for relational data
            natural graph processing

        Disadvantages
            computationally expensive
            complex implementation

        Applications
            social networks
            fraud detection
            recommendation systems
            molecular analysis

12. Radial Basis Function Network (RBFN)

        Neural network using radial basis functions.

        Structure:
            Input → RBF Hidden Layer → Output

        Working

        RBFN:
            measures distance from center points
            activates neurons based on proximity
            performs interpolation

        Features
            localized activation
            fast learning
            smooth approximation

        Advantages
            fast training
            good interpolation capability

        Disadvantages
            poor scalability
            sensitive to center selection

        Applications
            function approximation
            pattern recognition
            forecasting

13. Self-Organizing Map (SOM)

        Unsupervised neural network for clustering and visualization.

        Structure:
            Input → Competitive Layer → Map

        Working

        SOM:
            maps high-dimensional data
            groups similar data together
            creates lower-dimensional representation

        Features
            unsupervised learning
            clustering
            visualization capability

        Advantages
            good data visualization
            dimensionality reduction

        Disadvantages
            lower accuracy on complex tasks
            slower convergence

        Applications
            clustering
            visualization
            customer segmentation

14. Variational Autoencoder (VAE)

            Advanced probabilistic version of an autoencoder.

            Structure:
                Input → Encoder → Latent Space → Decoder → Output

            Working

            The VAE:
                encodes input into probability distributions
                samples latent vectors
                reconstructs data from latent space
                learns meaningful compressed representations

            Core Idea
                Instead of memorizing data directly, the VAE learns a smooth latent distribution for generating new data.

            Features
                probabilistic latent representation
                generative neural network
                continuous latent space
                supports data generation and reconstruction

            Advantages
                generates new realistic samples
                learns meaningful feature representations
                smooth latent space interpolation
                useful for unsupervised learning

            Disadvantages
                generated outputs may appear blurry
                training can be mathematically complex
                lower image sharpness than GANs

            Applications
                image generation
                anomaly detection
                data compression
                representation learning
                generative AI
                medical

# Padding and Masking  
    Padding 
        Adding extra values (usually zeros) to shorter sequences so every sequence becomes the same length.

        Why Padding is Needed
            Neural networks use matrix operations on GPUs.

        GPUs are optimized for: Fixed-size tensors

        Without padding:
            batching becomes difficult
            GPU parallelization becomes inefficient
            training becomes slow

        Types of Padding
            1. Post Padding (Most Common)

                Padding added at the end.

                Used in:
                    RNN
                    LSTM
                    GRU
                    Transformers

            2. Pre Padding

                Padding added at the beginning.

                Sometimes used in:
                    older sequence models
                    autoregressive tasks

        Problem Created by Padding

            Padding adds fake data.

            If the neural network learns from them:
                accuracy drops
                attention becomes wrong
                gradients become noisy

            This is where masking is used.

    Masking 
        Masking tells the neural network:  "Ignore these padded values."
        Why Masking is Important

            Without masking: The model may think that 0 = meaningful word

            This causes:
                    wrong attention scores
                    bad hidden states
                    poor predictions

            Masking prevents learning from fake tokens.
