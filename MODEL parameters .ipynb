1. LINEAR REGRESSION
    What Does It Do?
    Predicts continuous numerical values by finding a linear relationship between input features and output.

    How Does It Work?
        y = m₁x₁ + m₂x₂ + ... + mₙxₙ + b

        Where:
        - y = predicted value
        - m = coefficients/slopes
        - x = input features
        - b = intercept (y-axis crossing)

        Algorithm:
        1. Find the line that minimizes distance between predicted and actual values
        2. Use Ordinary Least Squares (OLS) to minimize: Σ(actual - predicted)²
        3. Solve for coefficients that minimize error

    Where to Use It?
        ✅ House price prediction (based on square footage, bedrooms, etc.)
        ✅ Sales forecasting (based on advertising spend)
        ✅ Temperature prediction (based on time of day)
        ✅ Stock prices (linear trend analysis)
        ✅ Salary prediction (based on experience, education)
        ✅ Demand forecasting
        ✅ Quick baseline models

    Advantages ✅
        Simple and fast
        Interpretable results
        Good baseline model
        Works well with linear relationships
    Disadvantages ❌
        Assumes linear relationship
        Poor with non-linear data
        Sensitive to outliers
        Limited flexibility


2.LOGISTIC REGRESSION
    What Does It Do?
    Predicts binary (2 classes) or multi-class outcomes using probability.

    How Does It Work?
        For Binary Classification:
        1. Transform linear combination through sigmoid function
        2. sigmoid(z) = 1 / (1 + e^(-z))
        3. Output probability between 0 and 1
        4. If probability > 0.5 → Class 1, else → Class 0

        For Multi-class:
        1. One-vs-Rest: Train multiple binary classifiers
        2. OR Multinomial: Single model with softmax function

        Cost Function (Log Loss):
        -1/m * Σ[y*log(ŷ) + (1-y)*log(1-ŷ)]

        Where:
        - y = actual value (0 or 1)
        - ŷ = predicted probability

    Where to Use It?
        ✅ Email spam detection (spam/not spam)
        ✅ Disease diagnosis (sick/healthy)
        ✅ Loan approval (approve/deny)
        ✅ Customer churn prediction (leave/stay)
        ✅ Fraud detection (fraud/legitimate)
        ✅ Customer sentiment (positive/negative/neutral)
        ✅ Credit risk assessment
        ✅ Patient mortality prediction

    Advantages ✅
        Fast and efficient
        Probability outputs
        Interpretable coefficients
        Works well for binary problems
    Disadvantages ❌
        Assumes linear boundary between classes
        Poor with non-linear separation
        Sensitive to imbalanced data
        Not flexible for complex patterns

4.Random Forest 
    What Does It Do?
    Combines multiple decision trees to make predictions more robust and accurate.

    How Does It Work?
        Step 1: Create Bootstrap Samples
                - Take random samples from data WITH replacement
                - Create N different training sets
                
        Step 2: Train Decision Trees
                - For each bootstrap sample, train a tree
                - Each tree can use different features
                - Trees can grow deep (no pruning)
                
        Step 3: Make Predictions
                For Classification:
                - Each tree votes on class
                - Choose class with most votes (majority voting)
                
                For Regression:
                - Average predictions from all trees

        Example with 3 trees:
        Tree 1 predicts: Class A
        Tree 2 predicts: Class A
        Tree 3 predicts: Class B
        → Final prediction: Class A (2/3 votes)

    Where to Use It?
        ✅ Feature importance ranking
        ✅ Banking (loan default prediction)
        ✅ Healthcare (disease prediction)
        ✅ E-commerce (recommendation systems)
        ✅ Image classification
        ✅ Stock market prediction

        ✅ General purpose classification/regression
        
    Advantages ✅
        Reduces overfitting (ensemble method)
        Handles non-linear relationships well
        Feature importance ranking
        Works with mixed data types
        Robust to outliers
        No feature scaling needed
        Often state-of-the-art
    Disadvantages ❌
        High memory usage
        Slow training on large datasets
        Less interpretable than single tree
        Biased toward dominant classes
        Many parameters to tune

7. GRADIENT BOOSTING (XGBoost)
    What Does It Do?
    Sequentially builds trees to correct mistakes of previous trees.

    How Does It Work?
        Code
        Step 1: Train first weak learner (tree)
                Tree 1 makes some predictions
                
        Step 2: Calculate residuals (errors)
                residual = actual - predicted
                
        Step 3: Train second tree on residuals
                Tree 2 learns to predict errors
                New prediction = Tree1 + learning_rate * Tree2
                
        Step 4: Repeat
                Tree 3 learns remaining errors
                Prediction = Tree1 + Tree2 + Tree3 + ...

        Example:
        Actual values:    [10, 20, 30]
        Tree 1 predicts:  [8, 18, 35]
        Residuals:        [2, 2, -5]
        Tree 2 predicts:  [2.5, 2.3, -4.8]
        Combined:         [8+2.5, 18+2.3, 35-4.8] = [10.5, 20.3, 30.2]
    Where to Use It?
        ✅ Kaggle competitions (winners often use XGBoost)
        ✅ Credit scoring
        ✅ Click-through rate prediction
        ✅ Customer churn prediction
        ✅ Sales forecasting
        ✅ Ranking problems
        ✅ General tabular data

    Advantages ✅
        Best performance on tabular data
        Handles non-linear relationships
        Feature importance ranking
        Works with mixed data types
        Robust to outliers
        State-of-the-art results
    Disadvantages ❌
        Complex hyperparameter tuning
        Risk of overfitting
        Slow training on large data
        High memory usage
        Less interpretable
        Requires careful validation

            ●   /      ●     \                       /

10. HIERARCHICAL CLUSTERING (DBSCAN)
    What Does It Do?
    Groups nearby points into clusters of arbitrary shape.

    How Does It Work?
        Step 1: Define epsilon (eps) - search radius
                - All points within eps distance are neighbors
                
        Step 2: Find core points
                - Points with >= min_samples neighbors
                
        Step 3: Form clusters
                - Connect core points that are neighbors
                - Add non-core points if close to core
                - Mark outliers (noise)

        Example:
        eps = 0.5, min_samples = 3

        ● ●●●          ← All 5 are core points
        ●●            ← All within 0.5 distance
        ● ●
            ○         ← Outlier (not core, far away)

        Result: One cluster with 7 points, 1 outlier

    Where to Use It?
        ✅ Anomaly detection
        ✅ Finding clusters of arbitrary shape
        ✅ Noise detection
        ✅ Spatial clustering
        ✅ Real-world complex patterns
