```python
from sklearn.linear_model import Lasso
import pandas as pd 
import numpy as np
data1 = pd.read_csv(r"handson-ml\datasets\housing\housing.csv") 
data1 = data1.dropna()
# removing the outliear
Q1 = data1['median_income'].quantile(0.25)
Q3 = data1['median_income'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
data1 = data1[(data1['median_income'] >= lower) & (data1['median_income'] <= upper)]
print("Columns and the data dictionary : ")
for i in range(len(data1.columns)):
        print(f"{data1.columns[i]} = {data1[data1.columns[i]].dtype}")
model1 = Lasso(alpha = 0.1 ) # lasso = Linear regression + L1 regularization 
x = data1[['median_income']] # Ml model expects the data to be in the 2d model
y = data1['median_house_value']

'''
Variable	Meaning
X_train	    Training inputs
X_test	    Testing inputs
y_train	    Training outputs
y_test	    Testing outputs

test_size=0.2 means 20% of the data is testing and rest is for training 
random_state=42 controls the randomness (splitting data , shuffling , random sampling , random initialization) and give the same values every spilt using this  
'''
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split( x, y,test_size=0.2,random_state=42)
from sklearn.model_selection import GridSearchCV
# Apha controls the penalty for the model 
params = {'alpha': [0.01, 0.1, 1, 10, 100] }
'''
Automatically find the best hyperparameters for a model.
estimator tells which model to work on 
param_grid used to set the paramter for the model 
cv is the cross validation , it sets the K values
scoring means how the values are evaluated , in this we are using r^2 
'''
grid = GridSearchCV( estimator=model1, param_grid=params, cv=5, scoring='r2' ) # method to choose hyperparameter (the setting and configration to set before training )
grid.fit(x_train,y_train)
# Best Hyperparameter
print("Best Parameter:")
print(grid.best_params_)
best_model = grid.best_estimator_
# Prediction
y_pred = best_model.predict(x_test)
# Evaluation Metrics
from sklearn.metrics import (mean_absolute_error,mean_squared_error,r2_score)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
# Print Results
print("\nMean Absolute Error:")
print(mae)
print("\nMean Squared Error:")
print(mse)
print("\nRoot Mean Squared Error:")
print(rmse)
print("\nR2 Score:")
print(r2)
```

Ridge Regression (L2 Regularization)

    Ridge adds the square of coefficients penalty.

    What Ridge does
        Shrinks coefficients toward zero
        Reduces model complexity
        Keeps all features
        Helps when features are highly correlated

    It usually:
        lowers variance,
        slightly increases bias,
        improves generalization.

    Choose Ridge when:

        1. Most features are useful

            Example:
                House price prediction
                Sensor data
                Economic dat
            You think every variable contributes a little.

        2. Multicollinearity exists
            When features are strongly correlated:
                height ↔ weight
                temperature ↔ humidity
            Ridge stabilizes coefficients.

        3. Number of features is large
            Especially when:
                many columns,
                but not necessarily sparse.

Lasso Regression (L1 Regularization)

    Lasso adds the absolute value penalty.

    What Lasso does
        Shrinks coefficients
        Can make some coefficients exactly 0
        Performs automatic feature selection
        This is the biggest difference.

    Choose Lasso when:

        1. You want feature selection
            Example:
                100 features but only 10 important
                Text classification
                Gene analysis
            Lasso removes irrelevant features automatically.

        2. Sparse datasets
            Where many features are unnecessary.

        3. Simpler interpretable models needed
            Because fewer features remain.
